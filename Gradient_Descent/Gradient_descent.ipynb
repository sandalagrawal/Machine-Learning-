```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples =4, n_features=1, n_informative=1, n_targets =1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# let's apply OLS first
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

**using gradient descent, we will try to achieve this intercept value - using the iterative approach.**

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color ='red')
```

```python
# let's apply Gradient Descent assuming slope is constant, m = 78.35
# and let's assume the starting value for intercept be b=0

y_pred = ((78.35 *X ) +0 ).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label = 'OLS')
plt.plot(X,y_pred, color ='#00a65a', label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

# here we are applying the formula to calculate slope.
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# let the learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) +b).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color ='#00a65a', label ='OLS')
plt.plot(X,y_pred1, color ='red', label = 'b = {}'.format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b=0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) +b ).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color ='#00a65a', label ='OLS')
plt.plot(X,y_pred2, color ='red', label = 'b = {}'.format(b))
plt.plot(X,y_pred1, color ='black', label = 'b = {}'.format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b=0')
plt.legend()
plt.show()
```

```python
# iteration 3

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) +b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color ='#00a65a', label ='OLS')
plt.plot(X,y_pred3, color ='red', label = 'b = {}'.format(b))
plt.plot(X,y_pred2, color ='#A3E4D7', label = 'b = {}'.format(b))
plt.plot(X,y_pred1, color ='black', label = 'b = {}'.format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b=0')
plt.legend()
plt.show()
```

```python
# looping this all

b = -100
m = 78.35
lr = 0.1

epochs =10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m*X + b

  plt.plot(X, y_pred)

plt.scatter(X,y)
```

**here, if the learning rate is very small, then it will take more time to reach the optimal line**

```python
# looping this all

b = -100
m = 78.35
lr = 0.01

epochs =10

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m*X + b

  plt.plot(X, y_pred)

plt.scatter(X,y)
```

**when we load date and time data into pandas, they get default data type as objects. In order to use the pandas function on them, we have to convert their data type to date and time explicitly.**

# **Working with Dates**

```python
# converting to datetime data type
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

**1. Extract Year**

```python
date['date_year'] = date['date'].dt.year

date.head()
```

**2. Extract month**

```python
date['date_month'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

**3. Extract days**

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week

date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns = ['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday' , 'Saturday']), 1,0)

date.drop(columns = ['product_id','city_id','orders']).head()
```

**Extract week of the year**

```python
date['date_week'] = date['date'].dt.week

date.drop(columns = ['product_id','city_id','orders']).head()
```

**Extract Quarter**

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns = ['product_id','city_id','orders']).head()
```

**Extract Semester**

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1 , 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

**Extract time elapsed between dates**

```python
import datetime

today = datetime.datetime.today()

today
```

```python
# subtracting date from today
today - date['date']
```

```python
# if we only want the difference in number of days

(today - date['date']).dt.days
```

```python
# months passed

np.round((today - date['date']) / np.timedelta64(1,'M'),0)
```

```python
# converting to datetime datatype

time['date'] = pd.to_datetime(time['date'])
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second
```

**Extract time**

```python
time['time'] = time['date'].dt.time

time.head()
```

```python
# find time difference

today - time['date']
```

```python
# in seconds

(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
